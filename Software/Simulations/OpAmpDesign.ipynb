```python
# design gain and offset as noted in 
# https://www.ti.com/lit/an/sloa097/sloa097.pdf?ts=1681808623443&ref_url=https%253A%252F%252Fwww.google.com%252F


#define needed parameters
Vref = 3.3 #reference voltage
Voutfs = -10
Voutzs = -3.7
Vinfs = 10
Vinzs = -10

#Calculate m and b
m = (Voutfs-Voutzs)/(Vinfs-Vinzs)
b = Voutzs - m*Vinzs

print('m: ', m)
print('b: ', b)
```

```python
#Negative m and negative b
#choose Rf
Rf = 10000

#Calculate Rg1 and Rg2
Rg1 = Rf/abs(m)
Rg2 = Vref*Rf/abs(b)

print('Rg1: ', Rg1)
print('Rg2: ', Rg2)
```

```python
#Negate m positive b
#choose 
Rf = 330000
R2 = 100000

#calculate
Rg = Rf/abs(m)
R1 = (b*R2*Rg)/(Vref*(Rf+Rg)-b*Rg)

print('Rg: ', Rg)
print('R1: ', R1)
```

```python
#Positive m and positive b
#this is only for greater than unity gain. not prefered circuit
R1 = 1
R2 = (Vref*R1*m)/b
Rf = 40
Rg = (R2*Rf)/(m*(R1+R2)-R2)

print('Rg: ', Rg)
print('R2: ', R2)
```
